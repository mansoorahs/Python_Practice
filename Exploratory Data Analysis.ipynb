```python
from numpy import *
import numpy as np
from scipy.integrate import *
from matplotlib.pyplot import *

x=np.array([0,1,2,3,4,5])
y=np.array([0,0.8,0.9,0.1,-0.8,-1])

print (x,y)

p1 = polyfit(x,y,1)

%matplotlib inline
plot(x,y,'o')


show()
```

```python
plot(x,y,'o')
plot(x,polyval(p1,x),'r-')
```

```python
p2 = polyfit(x,y,2)
p3 = polyfit(x,y,3)

print (p1)
print (p2)
print (p3)
```

```python
plot(x,y,'o')
plot(x,polyval(p1,x),'r-')
plot(x,polyval(p2,x),'b--')
plot(x,polyval(p3,x),'m:')

```

```python
plot(x,y,'o')
xp=linspace(-2,6,100)
plot(xp,polyval(p1,xp),'r-')
plot(xp,polyval(p2,xp),'b--')
plot(xp,polyval(p3,xp),'m:')
```

```python
%matplotlib inline
from pylab import *
import numpy as np

np.random.seed(4)
x=np.random.normal(3.0,1.0,1000)
y=np.random.normal(50.0, 10.0, 1000)/x

scatter (x,y)
```
